```python
import os
import pandas as pd
```

```python
path = 'in/'
```

```python
def group_file(file):
    # read sheet
    df = pd.read_excel(file)    
    B = 1
    O = 14
    P = 15
    # mean of O and P
    df_mean = df[{df.columns[B], df.columns[O],df.columns[P]}].groupby(df.columns[B]).mean()
    df_mean.reset_index(inplace=True)
    df_mean.columns = [df.columns[B],'avg_S', 'avg_T']
    # first of other columns
    df_first = df.drop([df.columns[O],df.columns[P]], axis=1)
    df_first = df_first.groupby(df.columns[B]).first()
    df_first.reset_index(inplace=True)
    # group
    df_grouped = df_first.merge(df_mean,on=df.columns[B])
    return df_grouped
```

```python
# read files
sheet = None
for file in os.listdir(path):
    if sheet is None:
        sheet = group_file(os.path.join(path, file))
    else:
        sheet = sheet.append(group_file(os.path.join(path, file)))
```

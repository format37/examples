Загружаем библиотеки

```python
from catboost import Pool, CatBoostRegressor
import pandas as pd
from sklearn.model_selection import train_test_split
from catboost.utils import eval_metric
```

Читаем данные из файла

```python
df = pd.read_csv('file_training.csv', sep=',', header = None)
```

Входные параметры

```python
X = df.drop(df.columns[0], axis=1)
X
```

Ответы

```python
y = df[df.columns[0]]
y
```

Разделяем датанные на обучающие и тестовые в случайном порядке

```python
X_train, X_validation, y_train, y_validation = train_test_split(X, y, train_size=0.75, random_state=42)
eval_dataset = Pool(data=X_validation, label=y_validation)
```

Настройка модели   
https://catboost.ai/en/docs/concepts/python-reference_catboostregressor

```python
model = CatBoostRegressor(
    boost_from_average=True,
    one_hot_max_size = 256,
    depth = 16,
    langevin = True,
    verbose=True,
    task_type="GPU"
    )
```

Обучение

```python
model.fit(X_train, y_train, use_best_model=True, eval_set=eval_dataset)
```

Оценка модели

```python
pred = model.predict(X_validation)
params = model.get_params()
response = str(params)+'\n'
response += '\n'+params['loss_function']+' loss: '+ str(eval_metric(y_validation.to_numpy(), pred, params['loss_function']))
response += '\nFitted: '+str(model.is_fitted())
response += '\nModel score:\n'+str(model.score(X,y))
response
```

Сохранение модели

```python
model.save_model('model.dat')
```

Загрузка модели

```python
model = CatBoostRegressor()
model.load_model('model.dat')
```

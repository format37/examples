```python
import os
import pandas as pd
```

```python
path = 'in/'
```

```python
columns_map = {
        'AFL':'F',
        'sinus':'S',
        'Afib':'A',
        'non.AFL':'NF',
        'non.sinus':'NS',    
        'non.Afib':'NA'    
    }
columns = []
```

```python
def group_file(file):    
    # read sheet
    df = pd.read_excel(file)    
    # mean of O and P
    df_grouped = df.groupby(df.columns[1]).mean()
    df_grouped.reset_index(inplace=True)    
    return df_grouped
```

```python
# read files
sheet = None
for file in os.listdir(path):
    filepath = os.path.join(path, file)
    if sheet is None:
        sheet = group_file(filepath)
    else:
        sheet = sheet.append(group_file(filepath))
```

```python
# rename columns
sheet.rename(columns=columns_map, inplace=True)
```

```python
# max
sheet = sheet.groupby(['Filename']).max()
sheet.reset_index(inplace=True)
```

```python
# set empty values for non-zero columns
blank_column = pd.DataFrame(['' for i in range(len(sheet))])
sheet['class3'] = blank_column
sheet['class2'] = blank_column
sheet['Afib'] = blank_column
sheet['non.Afib'] = blank_column
sheet['class2_actual'] = blank_column
sheet['class2_predicted'] = blank_column
```

```python
# reorder columns
sheet = sheet[[
    'Patient_ID',
    'Filename',
    'SDRR',
    'SDPR',
    'MPv',
    #'SDVm130',
    'MRR',
    'SDRR_ratio',
    'SDPR_ratio',
    'MPv_ratio',
    'SDVm130_ratio',
    'MRR_ratio',
    'class3',
    'class2',
    'Afib',
    'non.Afib',
    'class2_actual',
    'class2_predicted',
    'F', 'S', 'A', 'NF', 'NS', 'NA'
    ]]
```

```python
sheet.sort_values(['Patient_ID'], ascending = False)
```

```python
import pandas as pd
import sklearn
# import catboost
from matplotlib import pyplot as plt
from catboost import CatBoostRegressor
from catboost.utils import eval_metric
```

```python
# News
news = pd.read_csv('temp.csv')
news = news[['date','A bull market is expected','A bear market is expected']]
news.columns = ['date','bull','bear']
news['date'] = pd.to_datetime(news['date']).dt.date
news = news.sort_values('date')
news = news.groupby(['date']).mean()
news = news.reset_index()
news['date'] = news['date'] + pd.Timedelta(days=1)
news.head(1)
```

```python
# Btc price
btc = pd.read_csv('BTC_USD Bitfinex Historical Data_2018_07_24-2022_07_24.csv')
btc = btc[['Date','Change %']]
btc.columns = ['date','change']
btc['date'] = pd.to_datetime(btc['date']).dt.date
btc['change'] = btc['change'].str.replace('%', '')
btc['change'] = btc['change'].astype(float)
btc = btc.sort_values('date')
btc.head(1)
```

```python
# join news and btc by date
df = pd.merge(news, btc, on='date')
df.head()
```

```python
len(news), len(btc), len(df)
```

```python
# Train a catboost regressor to predict the change
# model = CatBoostRegressor(iterations=1000, learning_rate=0.1, depth=6, loss_function='RMSE', random_seed=42)
model = CatBoostRegressor(loss_function='RMSE', random_seed=42)
model.fit(df[['bull','bear']], df['change'])
```

```python
# evaluate model score
model.score(df[['bull','bear']], df['change'])
```

```python
# evaluate feature importance
feature_importance = model.get_feature_importance()
feature_importance
```

```python
# evaluate loss
prediction = model.predict(df[['bull','bear']])
eval_metric(df['change'], prediction, metric='RMSE')
```
